Plot number of completed studies each month in a given year – for the submission dataset, the year
is 2021. You need to include your visualization as well as a table of all the values you have plotted
for each month.

```python
df1['Completion'] = pd.to_datetime(df1['Completion'])

df1['Completion_Year'] = df1['Completion'].dt.year

df_2021 = df1[df1['Completion_Year'] == 2021]

monthly_counts = df_2021.groupby(df_2021['Completion'].dt.month).size()

monthly_counts_table = pd.DataFrame({'Month': monthly_counts.index, 'Count': monthly_counts.values})

plt.bar(monthly_counts.index, monthly_counts.values)
plt.xlabel('Month')
plt.ylabel('Number of completed studies')
plt.title('Completed Studies in 2021')
plt.show()

# display the table
print(monthly_counts_table)
```
